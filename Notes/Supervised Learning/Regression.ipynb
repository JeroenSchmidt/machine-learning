## Linear Regression 

Out of all the possible linear regression lines, this minimizsed the error the training data the best. 

**Linear Line**
$y=mx+b$

The question is, how do we minimize the error of this function?

Lets consider the case were we are finding a constant function:

We define the following
* training date $X,y$ 
* function we want to approximate as $f(x) = c$ 
* cost function as $E(c)$- in this case we are using the squared error

Because the defined cost function is continues, we can differentiate $E(c)$



Inorder to find the optimal solution all we need to do is find c when the derivative $\frac{dE(c)}{dc}$ is equal to 0.

**Cool side note:** The above equation then reduces to the mean!

## Errors

Training data has errors. We are not modeling $f$, but $f+\epsilon$. This is one of the reasons why we use squared error.

These errors can could come from:
* Sensor error
* Malicious data
* Transcription error
* Unmodeled influences
