# PCA - Principal Component Analysis

**What it does**
* Determines axis that maximizes variance (the red lines) 
 * (i.e. transformation into a new space)
* Creates othogonal axis to that
* Then look at eigenvalue of all new axis created and determine which dimensions we can discard according to some information loss tolerance.

**Properties**
* Move from N -> m dimensions
* global algorithm
* best reconstruction
 * min $L_2$ error
* eigenvalue ~ 0 -> ignorable
* well studies -> fast
